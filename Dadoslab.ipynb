```python
from __future__ import division, print_function
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

%matplotlib inline


#Unidades em segundos, metros e passos

pf= np.array([27.00, 22.00, 18.00]) #dados passos Felipe
pr= np.array([29.00, 24.00, 22.00]) #dados  passos Raul
pc= np.array([31.00, 28.00, 21.00]) #dados passos Camila
pa= np.array([31.00, 27.00, 20.00]) #dados passos André
tf= np.array([16.88, 11.90, 8.60]) # dados tempo felipe
tr= np.array([16.94, 12.66, 9.56]) #dados tempo raul
tc= np.array([19.82, 15.60, 9.15])#dados tempo camila
ta= np.array([18.54, 15.18, 8.97])#dados tempo andré

x= 15.16
# Funções

def V(t):
   return x/t
def Ca(p, t):
  return  p/t
def Cp(p):
  return x/p
#funções para aproximações
def func(x, a):
    return a*x

print('V(t)= Distância/tempo')
print('Cadence(p, t)= Passos / Tempo')
print('Comprimento de Passos(p)= Distância/Passos')
```

```python

#subplots
f, avp = plt.subplots(2,2, sharex=True, sharey=True, figsize=(20,10))
plt.suptitle('Velocidade x Comprimento do Passo', fontsize=20)
avp[0,0].set_ylabel('Velocidade $[m/s]$ ')
avp[0,1].set_ylabel('Velocidade $[m/s]$')
avp[1,0].set_ylabel('Velocidade $[m/s]$')
avp[1,1].set_ylabel('Velocidade $[m/s]$')
avp[0,0].set_xlabel('Comprimento do passo [Metros/passo]')
avp[0,1].set_xlabel('Comprimento do passo [Metros/passo]')
avp[1,0].set_xlabel('Comprimento do passo [Metros/passo]')
avp[1,1].set_xlabel('Comprimento do passo [Metros/passo]')

#Graficos Individuais



avp[0,0].plot(Cp(pa), V(ta), 'o', Cp(pa), V(ta), '--', color='b', linewidth='1')
avp[0,0].set_title('André', fontsize=14)


avp[0,1].plot(Cp(pc), V(tc), 'o', Cp(pc), V(tc), '--', color='b', linewidth='1')
avp[0,1].set_title('Camila', fontsize=14)


avp[1,0].plot(Cp(pr), V(tr), 'o', Cp(pr), V(tr), '--', color='b', linewidth='1')
avp[1,0].set_title('Raul', fontsize=14)

avp[1,1].plot(Cp(pf), V(tf),'o', Cp(pf), V(tf), '--', color='b', linewidth='1')
avp[1,1].set_title('Felipe', fontsize=14)

#fim graficos individuais

```
